## Some Bivariate and Multivariate analysis

### Profit Vs Discount & Profit Vs Quantity :

```python
plt.figure(figsize=(30,15))
plt.subplot(1,2,1)
sns.lineplot(x='Discount',y='Profit',label='Profit',data=retail)
plt.title('PROFIT VS DISCOUNT',fontsize=20)

plt.subplot(1,2,2)
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=retail)
plt.title('PROFIT VS QUANTITY',fontsize=20)
plt.legend()
plt.show()
```

* There are no correlation between discount and profit.

* Ther are some relation between Quantity vs profit.

### Segmentwise Sales and Profit

```python
retail.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['lightblue','blue'],figsize=(10,6))
plt.ylabel('Profit-Loss and sales')
plt.xlabel('Segments')
```

* Profit and loss according to the sale is high in consumer segment.

### region wise sales and profit

```python
retail.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['lightblue','blue'],figsize=(10,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

* Profit and Sales are high in the west region.
* Profit and sales is very low in the central and south region. 
* We have to focus on the central and south region why the sales and profit is low

### Region and segmentwise Sales and Profit

```python
plt.figure(figsize=(12,8))
sns.set_palette('coolwarm_r')
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=retail,hue='Segment',order=retail['Region'].value_counts().index)
plt.xlabel('Region',fontsize=15)
plt.show()
```

* region wise sales are almost same

### Statewise Sales and Profit

```python
retail.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False).plot(kind='bar',color=['blue','lightblue'],figsize=(14,6))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

* Profit & Sales both are high in California and New york

* Profit & Sales both are low in South west virginia and North Dekota

* There are loss in texas pennsylvania,ohio,llinosis,north california.

* We have to look into that states why the sales are so low we have to fix the problen to get   more profit.

```python
t_states = retail['State'].value_counts().nlargest(10)
t_states
```

### Profit-Loss vs Category

```python
retail.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['blue','lightblue'],alpha=0.9,figsize=(12,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

* Sales & Profit in the Technology category is very High rather than office supplies and furniture

* Team need to focused on the furniture category for more profit.

### Profit-Loss vs subcategory

```python
retail.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False).plot.bar(color=['blue','lightblue'],figsize=(15,8))
plt.title('Profit & loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

* Phones sub-category have high sales.
* chairs have high sales but less profit compared to phones
* Tables and Bookmarks sub-categories facing huge loss
